#Criando coluna horarios

```python
#Definindo os intervalos de hórarios

import numpy as np
conditions = [np.logical_and(tv_0_with_rat['Hora Início'].gt(0), np.less_equal(tv_0_with_rat['Hora Início'], 9.75)),
  np.logical_and(tv_0_with_rat['Hora Início'].gt(9.90), np.less_equal(tv_0_with_rat['Hora Início'],11.75)),
  np.logical_and(tv_0_with_rat['Hora Início'].gt(11.9), np.less_equal(tv_0_with_rat['Hora Início'],14.75)),
  np.logical_and(tv_0_with_rat['Hora Início'].gt(14.9), np.less_equal(tv_0_with_rat['Hora Início'],17.75)),
  np.logical_and(tv_0_with_rat['Hora Início'].gt(17.9), np.less_equal(tv_0_with_rat['Hora Início'],20.75)),
  np.logical_and(tv_0_with_rat['Hora Início'].gt(20.9), np.less_equal(tv_0_with_rat['Hora Início'],24.50)),
  np.logical_and(tv_0_with_rat['Hora Início'].gt(24.6), np.less_equal(tv_0_with_rat['Hora Início'],29.75)),
]
```

```python
#Array com os tipos de horários que teremos como base

outputs = ['Manha 1', 'Manha 2' , 'Almoço', 'Tarde', 'Noite 1', 'Noite 2', 'Madrugada']
```

```python
#Criando a coluna Horarios nas planilhas

tv_0_rat['Horarios'] = pd.Series(np.select(conditions, outputs))
tv_1_rat['Horarios'] = pd.Series(np.select(conditions, outputs))
tv_2_rat['Horarios'] = pd.Series(np.select(conditions, outputs))

```

```python
#Convertendo a coluna Data para apenas o ano 
tv_0_rat['Data'] = pd.to_datetime(tv_0_rat['Data']).dt.year
tv_1_rat['Data'] = pd.to_datetime(tv_0_rat['Data']).dt.year
tv_2_rat['Data'] = pd.to_datetime(tv_0_rat['Data']).dt.year
```
